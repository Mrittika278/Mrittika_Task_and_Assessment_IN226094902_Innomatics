```python
def smart_parking(capacity, logs):
    parked = 0
    peak = 0
    for log in logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1
        if parked > peak:
            peak = parked
    status = "Available"
    if parked > capacity:
        status = "Full"
    print("Currently Parked Vehicles:", parked)
    print("Parking Status:", status)

Parking_Capacity = 50
Vehicle_Logs =["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
smart_parking(Parking_Capacity,Vehicle_Logs)
```

```python
def estimate_delivery_time(distance, traffic, weather):
    base_time = distance * 5

    if traffic == "High":
        base_time += 15
    elif traffic == "Medium":
        base_time += 8
    elif traffic == "Low":
        base_time += 3

    if weather == "Rainy":
        base_time += 10
    elif weather == "Stormy":
        base_time += 20
    elif weather == "Cloudy":
        base_time += 5

    print("Estimated Delivery Time:", base_time, "minutes")


estimate_delivery_time(8, "High", "Rainy")
```

```python
def theatre_occupancy(total_seats, booked_seats):
    booked_count = len(booked_seats)
    occupancy = (booked_count / total_seats) * 100

    if occupancy == 100:
        status = "Housefull"
        suggestion = "No additional shows required"
    elif occupancy >= 75:
        status = "Almost Full"
        suggestion = "Consider opening additional shows"
    else:
        status = "Seats Available"
        suggestion = "No additional shows required"

    print("Occupancy:", int(occupancy), "%")
    print("Show Status:", status)
    print("Suggestion:", suggestion)


booked = [1] * 150
theatre_occupancy(200, booked)
```

```python
def classify_server_load(cpu_readings):
    total = 0
    for reading in cpu_readings:
        total += reading

    average = total / len(cpu_readings)

    if average < 50:
        status = "Normal"
    elif average <= 80:
        status = "Warning"
    else:
        status = "Critical"

    print("Average CPU Load:", int(average), "%")
    print("Server Status:", status)


classify_server_load([45, 60, 70, 85, 90])
```

```python
def classroom_resource_monitor(resource_usage):
    overused = []

    for resource, hours in resource_usage.items():
        if hours > 8:
            overused.append(resource)

    if overused:
        alert = "Yes"
    else:
        alert = "No"

    print("Overused Resources:", ", ".join(overused))
    print("Energy Alert:", alert)


usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

classroom_resource_monitor(usage)
```

```python
def manage_event_registrations(capacity, registrations):
    confirmed = 0
    waitlisted = 0

    for i in range(registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1

    if waitlisted > 0:
        status = "Closed"
    else:
        status = "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)


manage_event_registrations(100, 105)
```
