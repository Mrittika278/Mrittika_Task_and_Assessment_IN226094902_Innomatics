```python
numbers = [45, 22, 89, 10, 66]
max = numbers[0]
min = numbers[0]
for num in numbers:
    if num > max:
        max = num
    if num < min:
        min = num
print(max)
print(min)
```

```python
prices = [450, 1200, 899, 1500, 300]
count = 0
for i in prices:
    if i > 1000:
        count = count + 1
print(count)
```

```python
attendance = ["P", "P", "A", "P", "P"]
count = 0
for i in attendance:
    if i == "P":
        count = count + 1
total_days = len(attendance)
attn_percent = (count / total_days) * 100
print("Attendance Percentage : ",attn_percent)
```

```python
word = "PythonProgramming"
char_count = {}
for char in word:
    if char in char_count:
        char_count[char] += 1
    else:
        char_count[char] = 1
print(char_count)
```
