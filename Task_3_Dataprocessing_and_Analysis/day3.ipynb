```python
import string
sentence = "Buy mobile phone buy phone online"
translator = str.maketrans('','',string.punctuation)
cleaned_text = sentence.translate(translator)
words = cleaned_text.lower().split()
freq = {}
for word in words:
    if word in freq:
        freq[word] += 1
    else:
        freq[word] = 1
    if freq[word] > 1:
        print(word)




```

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
valid_sensor_readings_pair = []
for readings in sensor_readings:
    if readings % 2 == 0:
        valid = (sensor_readings.index(readings),readings)
        valid_sensor_readings_pair.append(valid)
print(valid_sensor_readings_pair)
        
```

```python
emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]
domains = []
for email in emails:
    domain = email.split("@")[1]
    domains.append(domain)
freq = {}
for domain in domains:
    if domain in freq:
        freq[domain] += 1
    else:
        freq[domain] = 1
print(freq)
total = len(emails)
for domain in freq:
    value = freq[domain]
    percentage = value/total * 100
    print("Percentage usage of ",domain,"is",percentage)

    
    
```

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]
tot = 0
for sale in sales:
    tot = tot + sale
n = len(sales)
avg = tot/n
threshold = ((30 / 100) * avg) + avg
for day_number,sale_value in enumerate(sales,start = 1) :
    if sale_value > threshold:
        print("Day",day_number,":",sale_value)
        
```
